```python
import warnings
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.image as img_reading
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
import matplotlib.pyplot as plt
warnings.simplefilter('ignore')
```

```python
img=img_reading.imread('buird.jpg')
```

```python
img.shape
```

```python
plt.imshow(img,);
```

```python
flattened_3c = (np.float32(img)).reshape(-1, 3)
```

```python
def elbow_method(data):
    k_range = np.arange(1, 11)
    distortions = []

    for k in k_range:
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)
        distortions.append(kmeans.inertia_)

    plt.plot(k_range, distortions, marker='o')
    plt.xlabel('Number of clusters')
    plt.ylabel('Distortion')
    plt.title('Elbow Method for Optimal K')
    plt.show()
```

```python
elbow_method(flattened_3c)
```

```python
optimal_k = 2  # Adjust this value based on the elbow plot

# Perform K-means clustering with the chosen K
kmeans = KMeans(n_clusters=optimal_k).fit(flattened_3c)
segmented_data = kmeans.cluster_centers_[kmeans.labels_]
print("---------------------------------------------------------")
print(f"the centers of clusters of reigons of image are {kmeans.cluster_centers_}")

print("---------------------------------------------------------")
# Reshape the segmented data back to the shape of the original image
segmented_image = segmented_data.reshape(img.shape).astype(np.uint8)
plt.imshow(img)
plt.show()
# Display the segmented image
plt.imshow(segmented_image)
plt.show()
pd.DataFrame({"labels":kmeans.labels_})
```

```python
img_reading.imsave("segmented_image.jpg",segmented_image)
```
